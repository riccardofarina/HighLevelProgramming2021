## Mutable and Immutable with function arguments

Nella 99% dei casi le funzioni modificano solo localmente gli argomenti, creando copie. Tuttavia quando si agisce su un elemento di un mutable object (list, dictionary...) senza ridefinirlo completamente, la modifca e come globale, anche fuori dalla funzione

Se x e immutable, la funzione crea una copia ma non modifica veramente x

```python
def modify(x):
    x += 2
    return x

x = 1
print(x)
print(modify(x))
print(x)

#per cambiare globalmente, usare global
```

Se x e mutabile invece (come agendo su una lista), eseguire la funzione modifica la variabile anche fuori

```python
def modify(x):
    x[0] = 999
    return x

x = [1,2,3]
print(x)
print(modify(x))

print(x) #x rimane modificata!
```

Tuttavia se la funzione non agisce sull'oggetto ma lo sostituisce (cambia tutta la lista), crea un copia e non lo moifica

```python
def no_modify(x):
    x = [4,5,6]
    return x

x = [1,2,3]
print(x)
print(no_modify(x))
print(x)
```

### Initialization of function arguments

Un comportamento controintuitivo si ha quando si inizializzano variabili mutable nelle funzioni

In questo codice, una lista vuota x[] viene creata implicitamente, e la funzione agisce su quella (mutable), a meno che non sia specificato diversamente

```python
def f(x = []):
    x.append(1)
    return x

#append modifica il singolo elemento, quindi cambia globalmente il mutable object. Ogni chiamata appende 1
print(f())
print(f()) 

#Funzione agisce sul parametro passato  
print(f(x = [9,9]))

#Funzione torna ad agire sulla lista che aveva inizializzato prima
print(f())
```

What actually happens: a new list is created once when the function is defined, and the same list is used in each successive call.
Check this [post](https://docs.python-guide.org/writing/gotchas/).

```python
#con None al posto della inizializzazione, poi inizializzo solo se l'elemento non esiste
def f(x = None):
    if x is None:
        x = []
    x.append(1)
    return x

print(f())
print(f())
print(f(x = [9,9,9]))
print(f())
print(f())
```

## Higher order functions

HOF: Funzioni che hanno anche altre funzioni come arguments come `map` e `filter`

### map

`map` applica la funzione che gli si passa a tutta la lista

```python
x = list(map(square, [1,2,3])) 
print(x)

# Nota: map in python 3 ritorna un iterator
for i in map(square, range(5)): print(i)
```

### filter

`filter` testa gli elementi di una lista con una funzione (boolean) e ritorna solo quelli che soddisfano la condizione.

```python
def is_even(x):
    return x % 2 == 0 #return the condition result

print(is_even(11))

print(list( filter(is_even, [1,2,3,4]) ))
```

```python
#HOF of HOF 
list(map(square, filter(is_even, range(5))))
```

### reduce

`reduce` riduce una lista a un risultato applicando un binary operators (+,-, *...) tra gli elementi consecutivamente. Ora si usano cicli for

```python
from functools import reduce

def my_add(x, y):
    print("Adding", x, "and", y)
    return x + y

# another implementation of the sum function
reduce(my_add, [1,2,3,4,5])
```

### zip

`zip` permette di combinare elementi di liste diverse, e svolgere operazioni tra liste diverse con elementi degli stessi indici. 

```python
xs = [1, 2, 3, 4]
ys = [10, 20, 30, 40]
zs = ['a', 'b', 'c', 'd', 'e'] #anche con lunghezze diverse

for x, y, z in zip(xs, ys, zs):
    #operazioni su x,y,z elementi corrispondenti nelle 3 liste 
    print(x, y, z)
```

```python
def sum_on_index(xs, f):
    return sum(map(f, xs)) #applica f agli elementi e somma la lista

xs = [1,3,57]
print(sum_on_index(xs, cube))
```

### Returning a function

Le funzioni possono ritornare funzioni:

```python
def make_logger(target):
    #Definisco una funzione logger che prende una stringa e la scrive nel file target
    def logger(data):
        with open(target, 'a') as f:
            f.write(data + '\n')
    return logger

#Imposto la funzione per scrittura nel nuovo file
foo_logger = make_logger('foo.txt')
print(type(foo_logger)) #foo_logger e una funzione che aspetta data in input

foo_logger('Hello')
foo_logger('World')
```

### Decorators

I decorators sono funzioni che "decorano" una funzione, quando viene eseguita, aggiungendo istruzioni prima e dopo. 
In altre parole sono funzioni "wrapper" che fanno da sfondo alla funzione eseguita

Per usarli, indicare con `@` prima della def della funzione

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_whee():
    print("Whee!")
    
#Quando say_whee e eseguita, viene incastonata nel decorator
say_whee()
```

Senza decorator, il codice sarebbe cosi

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

def say_whee():
    print("Whee!")

say_whee = my_decorator(say_whee)

say_whee()
```

## Recursive functions 

```python
#Fibonacci with recursion
def fib1(n):
    
    # base case
    if n==0 or n==1:
        return 1
    # recursive case
    else:
        return fib1(n-1) + fib1(n-2)


print(fib1(5)) #termine 4 della sequenza
print([fib1(i) for i in range(10)])
```

```python
# Fibonacci without recursion, more efficient

def fib2(n):
    
    a, b = 0, 1
    for i in range(1, n+1):
        a, b = b, a+b
    return b

print([fib2(i) for i in range(10)])
```

```python
#execution time di un'istruzione

%timeit fib1(20)
%timeit fib2(20)
```

La ricursione e circa ~ 100 volte meno efficiente

## Iterators

Gli iterators permettono di scorrere un dataset caricando un dato alla volta
Sono utilizzati per rendere piu efficienti le operazioni con dataset troppo grandi per la RAM.
Solitamente sono impliciti nei cicli for

`iter()` crea l'iterator, `next()` scorre al valore successivo.

```python
# Iterators can be created from sequences with the built-in function iter()
x = [1,2,3]
x_iter = iter(x)

print(next(x_iter))
print(next(x_iter))
print(next(x_iter))
#print(next(x_iter))
```

```python
# Most commonly, iterators are used (automatically) within a for loop
# which terminates when it encouters a StopIteration exception

x = [1,2,3]
x_iter = iter(x)

print(next(x_iter))
print(next(x_iter))

for x in x_iter:
    print("Valore:",x)
```

# Classes and Objects

In python si possono definire le classi e gli oggetti. Il python doc per le classi e [questo](https://docs.python.org/3.6/tutorial/classes.html)

```python
# Class definition
class Pet:
    
    #gli attributi della classe sono name e age
    name = None
    age = None
    
    # Class methods (funzioni della classe)
    
    # the "constructor"
    def __init__(self, name, age):  #definizione delle private variables che costituiscono l'oggetto della classe 
        self.name = name
        self.age = age
    
    def convert_age(self, factor):
        self.age *= factor
# End class definition

mydog = Pet("Lilly", 12) # Crea un oggetto della classe "Pet", specificando nome e eta
print(mydog.name, mydog.age)

#Modifico direttamente i parametri dell'oggetto creato (senza getter o setter)
mydog.age = 3
print(mydog.name,mydog.age)

#Eseguo il metodo della classe che agisce sul parametro age. Ora il valore e modificato
mydog.convert_age(2)
print(mydog.age)
```

E possibile anche fare inheritance delle classi, cioe incastonarle in altre classi. 

La classe `Dog` agisce su oggetti della classe `Pet` con parametri `name` e `age`

```python
# ineritance della classe Pet
class Dog(Pet):
    
    #parametri della classe dog
    species = "mammal"
    
    # Metodi della classe Dog
    def convert_age_dog(self):
        self.age *= 7
        
    def ok_species(self):
        self.species = "ok!"
    
        
puppy = Dog("toby", 10) #Definisco l'oggetto della classe Dog, si richiama il costruttore della classe Pet
print(puppy.name, puppy.age)

puppy.convert_age_dog() #richiamo metodo convert_age_dog della classe Dog e modifico l'eta
print(puppy.age)

#Modifico il parametro species con il metodo ok_species
print(puppy.species)
puppy.ok_species()
print(puppy.species)
```
